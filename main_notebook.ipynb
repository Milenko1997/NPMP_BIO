```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

from hill_functions import *
from models import *
```

```python
# common parameters

t_end = 200
t_span = (0, t_end)
N = 1000

# model parameters
alpha1 = 50     # protein_production
alpha2 = 49.36  # protein_production
alpha3 = 50     # protein_production
alpha4 = 49.54  # protein_production
delta1 = 1.93   # protein_degradation
delta2 = 0.69   # protein_degradation
Kd = 10.44      # Kd
n = 4.35        # hill
params_ff = (alpha1, alpha2, alpha3, alpha4, delta1, delta2, Kd, n)

input_states_SIPO = [50,0,50,50]
input_states_PISO = [[50,0,50,50]]

```

```python
y = [0, 50, 0, 50, 0, 50, 0, 50, 0, 50, 0, 50, 0, 50, 0, 50] #init state all zeros
Y0 = [0, 50, 0, 50, 0, 50, 0, 50, 0, 50, 0, 50, 0, 50, 0, 50] #init state all zeros
T = np.linspace(0, t_end, N) 

##################### SIPO call #################################

SIPO_register = SIPORegister()
solution_SIPO = solve_ivp(lambda t, y: SIPO_register.model(y, t, input_states_SIPO, params_ff), t_span, Y0)

output_states = solution_SIPO.y.T 
T_solution_SIPO = solution_SIPO.t 

######################### FIGURE SIPO ###########################
fig, axs = plt.subplots(4, 1, figsize=(10, 15))  
colors = ['blue', 'orange', 'green', 'red']

for i in range(4):
    axs[i].plot(T_solution_SIPO, output_states[:, i*4+2], label=f'Q{i+1}', color = colors[i])  # Plotting the q output of each flip-flop
    axs[i].plot(T_solution_SIPO, get_clock(T_solution_SIPO),  '--', linewidth=2, label="CLK", color='black', alpha=0.25)
    axs[i].set_title(f'State Q{i+1} Over Time')
    axs[i].set_xlabel('Time')
    axs[i].set_ylabel('State')
    axs[i].legend()

fig.suptitle('SIPO Register State Over Time', fontsize=16)
fig.savefig('result_plot_SIPO.png', dpi=300) 
```
